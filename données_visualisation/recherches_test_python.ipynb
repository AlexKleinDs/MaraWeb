```python
df = pd.read_csv("sport.csv", sep="|")
df
```

```python
# minuscules
for i in range(df.shape[0]):
    df.loc[i, 'pub'] = int(df.loc[i, 'pub'][0:4])
    df.loc[i, 'sport'] = df.loc[i, 'sport'].lower()
    df.loc[i, 'out'] = df.loc[i, 'out'].lower()
    df.loc[i, 'npi'] = df.loc[i, 'npi'].lower()
```

```python
#double masque pour garder que les articles où npi et out ont une valeur pour les deux
df1 = df[df['out']!='[]']
df2 = df1[df1['npi']!='[]']
df2.reset_index(inplace = True)
df2 = df2.drop('index', axis=1)
```

```python
df2
```

```python
# Crée et exporte un df 
k = pd.DataFrame()
k['annee']=''
for annee in df2[df2['sport'] == "['various']"]['pub'].value_counts().index:
    if annee == 2020:
        t = pd.concat([k, df2[df2['pub'] == annee]['sport'].value_counts()])
        t.fillna(value=annee, inplace=True)
    else:
        t = pd.concat([df2[df2['pub'] == annee]['sport'].value_counts(), t])
        t.fillna(value=annee, inplace=True)  
t.reset_index(inplace = True)
t.rename(columns={'index': 'sport', 0:'nb_article'}, inplace=True)
for i in range(t.shape[0]):
    t.loc[i, 'sport'] = t.loc[i, 'sport'][1:-1].lower()
t.to_csv(r'C:/Users/Lipto/anaconda3/Cours/MdW/sport_.csv')
```
